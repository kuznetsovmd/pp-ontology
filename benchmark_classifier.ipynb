#### IMPORTS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from env import env

from builders.classifiers.config import *
from builders.classifiers.wrapper import build_model

ENV = env()
ENV["use_cuda"] = False
CONFIGS = eval_config(**ENV)
print(CONFIGS)
```

### PLOTS

```python
plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = ["Times New Roman"]

store = []

for c in CONFIGS:
    model = build_model(**c["model_conf"])

    if not model:
        continue

    stats = model.stats_mem
    print(model.stats_mem)

    best = np.argmin([s['v_loss'] for s in stats])
    print(f'[{model.name}] best epoch: {best}, t_f1: {stats[best]["t_f1"]}, v_f1: {stats[best]["v_f1"]}, t_pr_auc: {stats[best]["t_pr_auc"]}, v_pr_auc: {stats[best]["v_pr_auc"]}')
    store.append([model.name, best, stats[best]["t_loss"], stats[best]["v_loss"], stats[best]["t_f1"], stats[best]["v_f1"], stats[best]["t_pr_auc"], stats[best]["v_pr_auc"]])

    t_loss = [s['t_loss'] for s in stats]
    v_loss = [s['v_loss'] for s in stats]
    t_f1 = [s['t_f1'] for s in stats]
    v_f1 = [s['v_f1'] for s in stats]
    t_pr_auc = [s['t_pr_auc'] for s in stats]
    v_pr_auc = [s['v_pr_auc'] for s in stats]

    plt.figure(figsize=(4,4))
    plt.grid()
    plt.plot(t_loss)
    plt.plot(v_loss)
    plt.legend(['train bce', 'validation bce'])
    plt.savefig(f'resources/figures/{model.name}-bce.png', dpi=400)
    plt.show()

    plt.figure(figsize=(4,4))
    plt.grid()
    plt.plot(t_f1)
    plt.plot(v_f1)
    plt.legend(['train f1', 'validation f1'])
    plt.savefig(f'resources/figures/{model.name}-f1.png', dpi=400)
    plt.show()

    plt.figure(figsize=(4,4))
    plt.grid()
    plt.plot(t_pr_auc)
    plt.plot(v_pr_auc)
    plt.legend(['train pr-auc', 'validation pr-auc'])
    plt.savefig(f'resources/figures/{model.name}-pr-auc.png', dpi=400)
    plt.show()

pd.DataFrame(store, columns=['model', 'epoch', 't_loss', 'v_loss', 't_f1', 'v_f1', 't_pr_auc', 'v_pr_auc']).to_csv('resources/classifiers.csv')
```
